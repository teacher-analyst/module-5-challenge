### Analysis

The purpose of the animal study was to compare the performance of Pymaceuticals’ drug of interest, Capomulin, against the other treatment regimens including Ramicane, Infubinol and Ceftamin. 248 mice were involved in the study of which 51% were male and 49% were female. 

The drug regimen with the lowest median tumor volume was Ramicane with Capomulin not far behind with a median tumor volume of 41.6 mm3. 

Mice specifically on the Capomulin regimen were analysed. The correlation coefficient calculated (0.84) is close to 1 indicating that there is a positive correlation between the weight of the mouse and the average tumor volume. Also,the volume of tumor of a specific decreased in size by 4.5 mm3bover the span of 45 days.  

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")


# Display the data table for preview
merged_df
```

```python
# Checking the number of mice.
merged_df['Mouse ID']. nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
merged_df.loc[merged_df.duplicated, 'Mouse ID']
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
merged_df.loc[merged_df["Mouse ID"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
drop_mouse="g989"
cleaned_df = merged_df[merged_df["Mouse ID"]!= drop_mouse]
cleaned_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_df['Mouse ID']. nunique()
```


## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

mean_tumor_vol=cleaned_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_tumor_vol=cleaned_df.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
var_tumor_vol=cleaned_df.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
std_tumor_vol=cleaned_df.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
sem_tumor_vol=cleaned_df.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

# Assemble the resulting series into a single summary DataFrame.
summary_df=pd.DataFrame({
            "Mean Tumor Volume" : mean_tumor_vol,
            "Median Tumor Volume" : median_tumor_vol,
            "Tumor Volume Variance" : var_tumor_vol,
            "Tumor Volume Std. Dev." : std_tumor_vol,
            "Tumor Volume Std. Err.": sem_tumor_vol
})
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_table2=cleaned_df.groupby("Drug Regimen").aggregate(['mean', 'median', 'var', 'std', 'sem'])["Tumor Volume (mm3)"]
summary_table2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = cleaned_df.groupby(["Mouse ID"])["Timepoint"].max()
max_timepoint = max_timepoint.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_max_df = max_timepoint.merge(cleaned_df, how="left", on=["Mouse ID", "Timepoint"])

merged_max_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Locate the rows which contain mice on each drug and get the tumor volumes

for drug in treatments:
    drug_data = merged_max_df.loc[(merged_max_df["Drug Regimen"] == drug)]
    final_tumor_vol = drug_data.loc[drug_data["Timepoint"] == drug_data["Timepoint"]]["Tumor Volume (mm3)"]
    tumor_vol.append(final_tumor_vol)    
    
 # Calculate the IQR and quantitatively determine if there are any potential outliers. 
for i,drug in enumerate(treatments):
    quartiles=tumor_vol[i].quantile([0.25, 0.5, 0.75]) 
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

 # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    outliers = tumor_vol[i][(tumor_vol[i] < lower_bound) | (tumor_vol[i] > upper_bound)]
    
    print(f"{drug}'s potential outliers:{outliers}")
   
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
red_out = dict(markerfacecolor='red',markersize=12)

ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xticklabels(treatments)
ax1.boxplot(tumor_vol, flierprops=red_out)

plt.show()
```

```python
#Filtering cleansed_df to only include Capomulin regimen
filtered_capomulin = cleaned_df[cleaned_df["Drug Regimen"] == "Capomulin"]

# Calculating average tumor volume of each mouse 
avg_tumor_vol = filtered_capomulin.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()

merge_tumor_weight = pd.merge(avg_tumor_vol, mouse_metadata, on='Mouse ID', how="left")

#plot scatterplot
plt.scatter(merge_tumor_weight["Weight (g)"], merge_tumor_weight["Tumor Volume (mm3)"], marker="o")

#Label axis
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

#set axis limits
plt.ylim(34, 47)
plt.xlim(14, 26)
plt.show()

```
